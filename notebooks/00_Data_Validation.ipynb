## Data Validation

### Objective

The purpose of this notebook is to validate the processed datasets generated by the preprocessing pipeline before performing Exploratory Data Analysis (EDA) and Machine Learning.

Only after successful validation will the datasets be used for EDA and model development.

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
PROJECT_ROOT = Path("..")
PROCESSED_DATA = PROJECT_ROOT / "data" / "processed"
INTERIM_DATA = PROJECT_ROOT / "data" / "interim"

print("Processed Data Folder:", PROCESSED_DATA)
print("Interim Data Folder:", INTERIM_DATA)
```

```python
resume_df = pd.read_csv(PROCESSED_DATA/"resumes_clean.csv")
jobs_df = pd.read_csv(PROCESSED_DATA/"jobs_clean.csv")
corpus_df = pd.read_csv(INTERIM_DATA/"job_corpus.csv")

print("Datasets loaded successfully!")
```

```python
print(f"Resume Dataset: {resume_df.shape}")
print(f"Jobs Dataset: {jobs_df.shape}")
print(f"Job Corpus: {corpus_df.shape}")
```

```python
print(resume_df.columns.tolist())

print(jobs_df.columns.tolist())

print(corpus_df.columns.tolist())
```

```python
# Duplicate Records

print(f"Resume Dataset : {resume_df.duplicated().sum()}")
print(f"Jobs Dataset   : {jobs_df.duplicated().sum()}")
print(f"Job Corpus     : {corpus_df.duplicated().sum()}")
```
